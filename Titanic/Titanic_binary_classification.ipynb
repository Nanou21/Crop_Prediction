```python
def cleaner(file):
    df = pd.read_csv(file)
    df = df.drop(columns =['PassengerId', 'Name','Cabin'], axis =1)
    df = df.dropna()
    objects = df.select_dtypes(exclude='number')
    mapper ={
    'TRAPPIST-1e':0,
    '55 Cancri e':1,
    'PSO J318.5-22':2
    }
    
    objects['Destination'].replace(mapper, inplace=True)
    home = {
    'Earth':0,
    'Europa':1,
    'Mars':2
    }
    objects['HomePlanet'].replace(home, inplace = True)
    objects['CryoSleep'] = objects['CryoSleep'].astype('int')
    objects['VIP'] = objects['VIP'].astype('int')
    numerical = new_titanic.select_dtypes(include='number')
    new = objects
    numeric = numerical.columns
    new[numeric] = numerical
    new.dropna(inplace = True)
#     X_scaled = scaler.fit_transform(new)

    X_t = new # X_scaled
#     X_t= X_t.dropna(inplace=True)
    return X_t
```
