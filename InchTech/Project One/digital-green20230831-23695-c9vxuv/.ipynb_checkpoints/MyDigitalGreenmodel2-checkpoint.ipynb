Checking Outliers using the a function that takes only values between the 0.1 and 0.9 quantile

```python
def cap_data(df):
    for col in df.columns:
        if (((df[col].dtype)=='float64') | ((df[col].dtype)=='int64')):
            low, high = df[col].quantile([0.01, 0.9])
            mask_area = df[col].between(low, high)
            df = df[mask_area]
        else:
            df[col]=df[col]
    return df

final_df=cap_data(df)
```
